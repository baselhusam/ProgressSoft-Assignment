```python
# Import Yolo
from ultralytics import YOLO

# Build the model
model = YOLO('yolov8n.pt')
```

#### Train the Model

the training using the `model.train` method, the hyperparameters that I specefied are:

* `path` : Path for the `yaml` file which include the number of classes, the name of it, and the training, val, and test data path.
* `epochs` : number of epochs, 100 in this case
* `optimizer` : the `Adam` optimizer in this case
* `batch` : 16 in this case

```python
model.train(data = '/content/drive/MyDrive/progressSoft/yolov8n.yaml',epochs = 100, optimizer= 'Adam', batch=16)
```
